```python
#movie ticket pricing system problem
def calculate_ticket_price(age,movie_type,show_time,day,is_student=False,group_size=1):
    
    if 0<=age<=12:
        base_price=8
    elif 13<=age<=17:
        base_price=10
    elif 18<=age<=64:
        base_price=15
    elif age>=65:
        base_price=12
    else:
        return "Invalid age"

    if movie_type=="regular":
        extra=0
    elif movie_type=='3D':
        extra=5
    elif movie_type=='IMAX':
        extra=8
    elif movie_type=='3D IMAX': 
        extra=10
    else:
        return "Invalid movie type"  
    
    
    total_price = base_price + extra
    
   
    if show_time=='Matinee':
        total_price*=0.80
    elif show_time=='Evening':
        pass
    elif show_time=='late night':
        total_price*=0.85
    else:
        return "Invalid show_time"
    
    if day in ['saturday','sunday']:
        total_price+=3
    
    if is_student and 13<=age<=64:
        total_price*=0.90
    
    final_total = total_price * group_size
    
    if group_size>=5:
        final_total*=0.85
    elif group_size>=10:
        final_total*=0.75
    
    return round(final_total,2)

price=calculate_ticket_price(age=25,
                            movie_type='IMAX',
                            show_time='Matinee',
                            day='saturday',
                            is_student=True,
                            group_size=5)
print('Total ticket price:', price)
```
